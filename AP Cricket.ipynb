```python
#function for defining team batting first

def bat1(t_sr):
    score = 0
    wickets = -1
    bpi = 0
    
    runs = [] #batsman ka runs scored in this innings
    balls = [] #balls faced by batsmen in this innings
  
    #120 balls in an innings    
    for sr in t_sr:
        if(((wickets<10)==True) and ((bpi <= 120)==True)):    
            ball = deliveries(bpi)
            
            if((ball + bpi) > 120):
                ball = 120 - bpi 
                out = False
            else:
                out = True 
            
            i_score = int( (sr/100)*ball)  #runs scored = (strike rate/100) * balls faced
            
            
            runs.append(i_score)
            balls.append(ball)
        
            
            bpi = bpi + ball #balls played in the innings by the team batting
            score = score + i_score #team score
            if(out==True):
                wickets = wickets + 1 
            
    return score,runs,balls,wickets  #runs,balls is list and score,wickets is int
```
